2.分析の実行

```python
# 目的変数
# 魅力度 = Attractive
y = df['魅力度']
# 説明変数（その他）
X = df.drop(columns=['魅力度'])
# 定数例の参加
import statsmodels.api as sm
X_ = sm.add_constant(X)

# モデルの定義
model = sm.OLS(y,X_)

# 分析の実行
result = model.fit()

# 結果の表示
summary= result.summary()
summary
```

```python
html = "<div>" + summary.as_html() + "</div>"
with open("好きなファイル名.html","w",encoding="utf-8") as f:
    f.write(html) 
```

```python
import pickle
with open("好きなファイル名.pkl", "wb") as f:
    pickle.dump(result,f)
```
